```python
def un_gz(file_name):#单个解压
    
    # 获取文件的名称，去掉后缀名
    f_name = file_name.replace(".gz", "")
    # 开始解压
    g_file = gzip.GzipFile(file_name)
    #读取解压后的文件，并写入去掉后缀名的同名文件（即得到解压后的文件）
    open(f_name, "wb+").write(g_file.read())
    g_file.close()
    
un_gz('E:/Data/PIOP1/sub-0001/anat/sub-0001_T1w.nii.gz')
def batch_un_gz(file_dir,file_name):#批量解压
    for i in file_name:
        un_gz(str(file_dir / f'{i}/anat'/ f'{i}_T1w.nii.gz'))
        print(str(file_dir / f'{i}/anat'/ f'{i}_T1w.nii.gz')+"   解压成功")
data_img_dir=Path('E:/Data/PIOP1')
batch_un_gz(data_img_dir,data_id)   
```

```python
smoothed_img = smooth_img("E:/Data/PIOP1/sub-0015/anat/sub-0015_T1w.nii", fwhm=3)
smoothed_array=np.array(smoothed_img.dataobj)
smoothed_array[44][55]
```

```python
def save_spatial_map(dir_dataset,ids,dataset_type):

    spatial_map = np.zeros((len(ids), 240, 240, 220), dtype=np.float32)

    for i, id in enumerate(ids):
        spatial_map[i]=smooth_img(str(dir_dataset / f'{id}/anat'/ f'{id}_T1w.nii'), fwhm=3).dataobj
        

    np.save(dir_dataset/f'input/spatial_map_{dataset_type}.npy', spatial_map)
```

```python
dir_dataset = Path('E:/Data/PIOP1')
save_spatial_map(dir_dataset, data_id.head(50), 'train')
```
